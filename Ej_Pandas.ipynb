## 1. Importar Pandas

Pandas es la librería de Python para análisis de datos. Su estructura principal es el **DataFrame**, similar a una tabla de Excel.

```python
import pandas as pd
import numpy as np

# Verificar versión de Pandas
print(f"Versión de Pandas: {pd.__version__}")
```

## 2. Crear DataFrames

Un DataFrame es una tabla bidimensional con filas y columnas etiquetadas.

```python
# Crear DataFrame desde un diccionario
datos = {
    'Estudiante': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Esteban'],
    'Edad': [20, 22, 19, 21, 23],
    'Promedio': [8.5, 7.2, 9.1, 8.8, 7.9],
    'Asistencia': [0.95, 0.80, 1.00, 0.90, 0.75]
}

df = pd.DataFrame(datos)
print("DataFrame:")
print(df)
print(f"\nTipo de dato: {type(df)}")
```

```python
# Crear DataFrame desde un array NumPy
datos_array = np.random.randn(4, 3)
df_array = pd.DataFrame(
    datos_array,
    columns=['Característica_A', 'Característica_B', 'Característica_C'],
    index=['Muestra_1', 'Muestra_2', 'Muestra_3', 'Muestra_4']
)
print("DataFrame desde array NumPy:")
print(df_array)
```

Antes de procesar datos, debemos entender su estructura, tipos y calidad.

```python
# Ver las primeras filas
print("Primeras filas (head):")
print(df.head())

# Ver las últimas filas
print("\nÚltimas filas (tail):")
print(df.tail(2))
```

```python
# Información del DataFrame
print("Información del DataFrame (info):")
print(df.info())

# Forma del DataFrame (filas, columnas)
print(f"\nForma: {df.shape}")
print(f"Filas: {df.shape[0]}, Columnas: {df.shape[1]}")
```

```python
# Estadísticas descriptivas
print("Estadísticas descriptivas (describe):")
print(df.describe())

# Estadísticas personalizadas
print("\nPromedio de Promedio:", df['Promedio'].mean())
print("Desviación estándar de Asistencia:", df['Asistencia'].std())
print("Valor máximo de Edad:", df['Edad'].max())
print("Valor mínimo de Promedio:", df['Promedio'].min())
```

## 4. Acceso a Datos: Filas y Columnas

Aprender a acceder a datos específicos es crucial para el procesamiento.

```python
# Acceder a una columna
print("Columna 'Edad':")
print(df['Edad'])
print(f"Tipo: {type(df['Edad'])}  # Es una Series")

# Acceder a múltiples columnas
print("\nColumnas 'Estudiante' y 'Promedio':")
print(df[['Estudiante', 'Promedio']])
```

## 5. Filtrado y Selección Condicional

Frecuentemente necesitamos seleccionar datos que cumplen condiciones específicas.

```python
# Estudiantes con promedio >= 8.5
print("Estudiantes con promedio >= 8.5:")
filtro = df['Promedio'] >= 8.5
print(df[filtro])

# Estudiantes jóvenes con buena asistencia
print("\nEstudiantes con edad < 21 y asistencia > 0.85:")
filtro_multiple = (df['Edad'] < 21) & (df['Asistencia'] > 0.85)
print(df[filtro_multiple])
```

```python
# Usar isin() para valores específicos
print("Estudiantes que son Ana o Diana:")
print(df[df['Estudiante'].isin(['Ana', 'Diana'])])

# Usar not con ~
print("\nEstudiantes cuyo promedio NO está entre 8.0 y 9.0:")
print(df[~((df['Promedio'] >= 8.0) & (df['Promedio'] <= 9.0))])
```

## 6. Operaciones de Agregación (GroupBy)

Agrupar datos por categorías permite extraer insights.

```python
# Crear un DataFrame con categorías
datos_categorias = {
    'Estudiante': ['Ana', 'Bruno', 'Carlos', 'Diana', 'Esteban', 'Fiona', 'Gastón'],
    'Grupo': ['A', 'B', 'A', 'B', 'A', 'B', 'A'],
    'Promedio': [8.5, 7.2, 9.1, 8.8, 7.9, 8.3, 9.0]
}
df_grupos = pd.DataFrame(datos_categorias)

# Agrupar por Grupo y calcular promedio
print("Promedio por grupo:")
print(df_grupos.groupby('Grupo')['Promedio'].mean())
```

```python
# Múltiples agregaciones
print("Estadísticas por grupo:")
print(df_grupos.groupby('Grupo')['Promedio'].agg(['mean', 'std', 'min', 'max', 'count']))
```

## 7. Manejo de Valores Faltantes (NaN)

Los datos reales frecuentemente tienen valores faltantes. Debemos identificarlos y manejarlos.

```python
# Crear DataFrame con valores faltantes
datos_faltantes = {
    'ID': [1, 2, 3, 4, 5],
    'Edad': [25, np.nan, 30, 28, np.nan],
    'Salario': [50000, 60000, np.nan, 55000, 65000],
    'Experiencia': [3, 5, np.nan, 4, np.nan]
}
df_faltantes = pd.DataFrame(datos_faltantes)

print("DataFrame con valores faltantes:")
print(df_faltantes)

# Detectar valores faltantes
print("\nValores faltantes (isna):")
print(df_faltantes.isna())
```

```python
# Contar valores faltantes
print("Número de valores faltantes por columna:")
print(df_faltantes.isna().sum())

# Porcentaje de valores faltantes
print("\nPorcentaje de valores faltantes:")
print((df_faltantes.isna().sum() / len(df_faltantes) * 100).round(2))
```

```python
# Opciones para manejar valores faltantes

# Opción 1: Eliminar filas con valores faltantes
df_sin_faltantes = df_faltantes.dropna()
print("Después de dropna():")
print(df_sin_faltantes)

# Opción 2: Rellenar con un valor constante
df_rellenado_media = df_faltantes.fillna(df_faltantes.mean())
print("\nDespués de fillna con media:")
print(df_rellenado_media)
```

```python
# Opción 3: Forward fill (propagación hacia adelante)
df_ffill = df_faltantes.fillna(method='ffill')
print("Después de forward fill:")
print(df_ffill)
```

## 8. Normalización y Estandarización (CRUCIAL para Redes Neuronales)

Las redes neuronales funcionan mejor cuando los datos están en un rango similar. Las dos técnicas principales son:

- **Normalización (Min-Max)**: Escala valores a [0, 1]
  - Fórmula: x_norm = (x - min) / (max - min)

- **Estandarización (Z-score)**: Transforma datos a media 0 y desviación estándar 1
  - Fórmula: x_std = (x - media) / desv_est

```python
# Datos de ejemplo para normalizar
datos_escalado = {
    'Ingresos': [30000, 50000, 80000, 120000, 45000],
    'Edad': [25, 32, 45, 55, 28]
}
df_escalado = pd.DataFrame(datos_escalado)
print("DataFrame original:")
print(df_escalado)
print(f"\nRangos: Ingresos [{df_escalado['Ingresos'].min()}, {df_escalado['Ingresos'].max()}], Edad [{df_escalado['Edad'].min()}, {df_escalado['Edad'].max()}]")
```

```python
# Normalización (Min-Max)
def normalizar(columna):
    return (columna - columna.min()) / (columna.max() - columna.min())

df_normalizado = df_escalado.copy()
df_normalizado['Ingresos'] = normalizar(df_escalado['Ingresos'])
df_normalizado['Edad'] = normalizar(df_escalado['Edad'])

print("Datos normalizados [0, 1]:")
print(df_normalizado)
print(f"\nRangos: Ingresos [{df_normalizado['Ingresos'].min():.2f}, {df_normalizado['Ingresos'].max():.2f}], Edad [{df_normalizado['Edad'].min():.2f}, {df_normalizado['Edad'].max():.2f}]")
```

```python
# Estandarización (Z-score)
def estandarizar(columna):
    return (columna - columna.mean()) / columna.std()

df_estandarizado = df_escalado.copy()
df_estandarizado['Ingresos'] = estandarizar(df_escalado['Ingresos'])
df_estandarizado['Edad'] = estandarizar(df_escalado['Edad'])

print("Datos estandarizados:")
print(df_estandarizado)
print(f"\nMedia de Ingresos: {df_estandarizado['Ingresos'].mean():.10f} (≈ 0)")
print(f"Desv. Est. de Ingresos: {df_estandarizado['Ingresos'].std():.2f} (≈ 1)")
```

## 9. Crear y Modificar Columnas

Frecuentemente necesitamos crear nuevas características (features) basadas en datos existentes.

```python
# Crear nueva columna basada en operación
df_temp = df.copy()
df_temp['Categoría'] = df_temp['Promedio'].apply(lambda x: 'Excelente' if x >= 9.0 else ('Bueno' if x >= 8.0 else 'Regular'))

print("DataFrame con nueva columna 'Categoría':")
print(df_temp)

# Crear columna numérica a partir de condición
df_temp['Aprobado'] = (df_temp['Promedio'] >= 7.0).astype(int)
print("\nDataFrame con columna 'Aprobado' (0=No, 1=Sí):")
print(df_temp[['Estudiante', 'Promedio', 'Aprobado']])
```
